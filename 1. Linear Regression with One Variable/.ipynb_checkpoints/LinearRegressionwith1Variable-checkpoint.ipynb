#### Hypothesis

```python
def hypothesis(X,theta):
    hyp=np.dot(X,theta)
    return hyp
```

#### Cost/Error Function

```python
def Cost(X, Y,  theta):
    N = Y.size
    cost = 0.0
    hyp=hypothesis(X,theta)
    cost = (1/(2*float(N)))*(np.sum(np.square(Y-hyp)))
    return(cost)
```

#### Gradient Descent

```python
def GradientDescent(X, Y, theta, alpha=0.0000001, num_iters=20):
    N = Y.size
    history = np.zeros(num_iters)
    for i in np.arange(num_iters):
        h = hypothesis(X,theta)
        theta = theta - alpha * ( (-1/float(N)) * (X.T.dot(Y-h)) )
        history[i] = Cost(X, Y, theta)
    return(theta, history)
```

We can evaluate using Sum of squared error or Root mean Squared error or any other type of error

We can instead use builtin library of sklearn that does everything for us. We can simply call function and pass in data. Lets see sklearn Linear Regression Model 

```python
from sklearn.linear_model import SGDRegressor
```

```python
Xtrain=np.expand_dims(np.array(train_data['sqft_living']),axis=1)
Ytrain=np.array(train_data['price'])
                
Xtest=np.expand_dims(np.array(test_data['sqft_living']),axis=1)
Ytest=np.array(test_data['price'])
```

```python
print(Xtrain.shape)
print(Ytrain.shape)

print(Xtest.shape)
print(Ytest.shape)
```

```python
model = SGDRegressor(n_iter=20,eta0=0.0000001,verbose=1)
model.fit(Xtrain, Ytrain)
```

Like error function, here we have builtin function named mean_squared_error

```python
from sklearn.metrics import mean_squared_error
```

```python
predictions = model.predict(Xtrain)
error= mean_squared_error(Ytrain.squeeze(), predictions)
print(np.sqrt(error) )
```
